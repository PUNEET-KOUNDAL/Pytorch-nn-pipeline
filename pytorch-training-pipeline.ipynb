```python
import numpy as np
import pandas as pd
import torch 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
df.shape
```

```python
X_train , X_test , y_train , y_test = train_test_split(df.iloc[ : , 1: ], df.iloc[:,0] , test_size=0.2)
X_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
y_train
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#transformning numpy array to pytorch tensor
```

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor
```

```python
y_train_tensor.shape
```

```python
#nueral network
X_train_tensor.shape[1]
```

```python
class mysimpleNN():
    def __init__(self,X):
        self.weight = torch.rand(X.shape[1],1,dtype=torch.
                                 float64 , requires_grad=True)
        self.bias = torch.rand(1,dtype=torch.float64,requires_grad = True)
        
    def forward(self,X,):
        z=torch.matmul(X,self.weight) + self.bias
        y_pred = torch.sigmoid(z)
        return y_pred

    
    def binary_cross_entropy(self,y_pred,y):
        loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
        return loss




        
        

```

```python
#important parameters
```

```python
learning_rate = 0.1
epochs =25
```

```python
model = mysimpleNN(X_train_tensor)
for epoch in range(epochs) :
    y_pred = model.forward(X_train_tensor)
    
    loss = model.binary_cross_entropy(y_pred,y_train_tensor)
    print(f'epoch: {epoch+1}, loss: {loss.item()}')

    
    loss.backward()
    with torch.no_grad():
        model.weight -= learning_rate * model.weight.grad
        model.bias -= learning_rate * model.bias.grad

    
    model.weight.grad.zero_()
    model.bias.grad.zero_()
    print(f'epoch {epoch+1},loss:{loss.item()}')

    
```

```python
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred > 0.9).float()
    
print(y_pred)
    
```

```python
accuracy = (y_pred == y_test_tensor).float().mean()
print(f'accurac: {accuracy.item()}')
```
